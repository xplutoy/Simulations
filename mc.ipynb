# MC: 蒙特卡洛方法
## 1. simple rejection sampling

```python
from functools import partial

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
%matplotlib inline 
```

```python
# unnormal p_t
f_t = lambda z: 0.3 * np.exp(-0.2 * z ** 2) + 0.7 * np.exp(-0.2 * (z - 10) ** 2)
q_z = partial(norm.pdf, loc=7, scale=5)
M = 12
X = np.arange(-5, 15, 0.1)
plt.plot(X, f_t(X))
plt.plot(X, M*q_z(X))
```

```python
rj, ac = 0, 0  #count
samples = []
for _ in range(1000000):
    u = np.random.uniform()
    x = norm.rvs(loc=7, scale=5)
    if u < f_t(x) / M * q_z(x):
        ac += 1
        samples.append(x)
    else:
        rj += 1

l_ = len(samples)
histogram, bins = np.histogram(samples, 20)
bins = 0.5*(bins[1:] + bins[:-1])
plt.plot(bins, histogram)
print('total {} ac {} rj {}'.format(100000, ac, rj))
```

*Notes*:
> 1. 接受率很低，受维度灾难影响大，无法扩展到高维。
> 2. q_z 必须是一个能容易采样的概率分布， p_z 可以不用归一化

## 2. simple importance sampling

### p_z和q_z都是概率分布

```python
p_z = partial(norm.pdf, loc=5, scale=3)
q_z = partial(norm.pdf, loc=0, scale=2)
f = lambda x: x
X = np.arange(-5, 11, 0.1)
plt.plot(X, p_z(X), color='red', label='p_z')
plt.plot(X, q_z(X), color='blue', label='q_z')
# plt.plot(X, f(X), color='green', label='f_z')
plt.legend(loc='upper left')
```

*评估期望*$\mathcal{E}[f]$

```python
L = 1000000
# 从q_z中采L个样本,loc=0, scale=2
q_samples = norm.rvs(size=L, loc=0, scale=2)
q_q = q_z(q_samples)
p_q = p_z(q_samples)
f_z = f(q_samples)
w = p_q / q_q
e_f_p = np.mean(w * f_z)

print('E[f]: {}'.format(e_f_p))
```

*Notes:*
> 1. 为保证采样的有效性，p_z 和 q_z 的分布应该有‘充分’的重叠区域。
> 2. 重要性采样本身并没有提供从p_z中采样的方法。
> 3. 对维度灾难敏感

*从p_z中采样：sampling-importance-resampling*
> stage1: $z1, ... , zL \sim q_z$  
> stage2: constructed w1, ... ,wL  
> stage3: choice(z, w, replace=True)  

```python
q_resamples = np.random.choice(q_samples, len(q_samples), True, w/np.sum(w))
hist1, bins1 = np.histogram(q_samples, 20)
bins1 = 0.5 * (bins1[1:] + bins1[:-1])
plt.plot(bins1, hist1)
hist2, bins2 = np.histogram(q_resamples, 20)
bins2 = 0.5 * (bins2[1:] + bins2[:-1])
plt.plot(bins2, hist2)

e_fp = np.mean(q_resamples)
v_fp = np.sqrt(np.var(q_resamples))
print('E[f]: {}, Var: {}'.format(e_fp, v_fp))
```

### p_z是非归一化的概率， q_z是归一化的概率分布

```python
# 用f_z表示未归一化的概率
f_z = lambda z: 0.3 * np.exp(-0.2 * z ** 2) + 0.7 * np.exp(-0.2 * (z - 5) ** 2)
q_z = partial(norm.pdf, loc=5, scale=2)
X = np.arange(-5, 10, 0.1)
plt.plot(X, f_z(X), color='red', label='f_z')
plt.plot(X, q_z(X), color='blue', label='q_z')
plt.legend(loc='upper left')
```

```python
L = 1000000
# 从q_z中采L个样本,loc=0, scale=2
q_samples = norm.rvs(size=L, loc=5, scale=2)
q_q = q_z(q_samples)
p_q = f_z(q_samples)
r = p_q / q_q   # 未归一化的概率比用r表示
```

估计p_z的配分函数$\frac{Z_p}{Z_q} \simeq \frac{1}{L}\sum_{l=1}^{L}r_l$, 由于q_z我们取的概率分布，所以$Z_q=1$. 

```python
Z_p = np.mean(r)
print('配分函数Z_p：{}'.format(Z_p))
```

估计p_z的期望:

```python
w = r / np.sum(r)
f = lambda x: x
E_p = np.sum(w * f(q_samples))
print('期望E_p: {}'.format(E_p))
```

*importance resampling*

```python
q_resamples = np.random.choice(q_samples, len(q_samples), True, w/np.sum(w))
hist1, bins1 = np.histogram(q_samples, 20)
bins1 = 0.5 * (bins1[1:] + bins1[:-1])
plt.plot(bins1, hist1, label='q_samoles')
hist2, bins2 = np.histogram(q_resamples, 20)
bins2 = 0.5 * (bins2[1:] + bins2[:-1])
plt.plot(bins2, hist2, label='q_resamples')
plt.legend()

e_fp = np.mean(q_resamples)
v_fp = np.sqrt(np.var(q_resamples))
print('E[f]: {}, Var: {}'.format(e_fp, v_fp))
```

## 3. Metropolis-Hastings

```python
def MH(n_samples, f_z, t_rvs, t_pdf):
    def T_(x, iter=250):
        rj = 0
        for _ in range(iter):
            x_pre = t_rvs(x)
            a = f_z(x_pre) * t_pdf(x, x_pre) / f_z(x) / t_pdf(x_pre, x)
            u = np.random.uniform()
            if u < a:
                x = x_pre
            else:
                rj = rj + 1
        return x, rj

    samples, rjs = [], []  # 样本和每次采样被拒绝的次数
    for _ in range(n_samples):
        x0 = t_rvs(0)
        x, rj = T_(x0)
        samples.append(x)
        rjs.append(rj)
    return samples, rjs

def test_MH(n_samples, f_z, t_rvs, t_pdf, label):
    samples, rjs = MH(n_samples, f_z, t_rvs, t_pdf)
    E_p = np.mean(samples)
    V_p = np.sqrt(np.var(samples))
    print('期望E_p：{:.3} 方差V_p: {:.3} 平均拒绝数rjs: {}'.format(E_p, V_p, np.mean(rjs)))
    hists, bins = np.histogram(samples, 60)
    bins = 0.5 * (bins[1:] + bins[:-1])
    plt.plot(bins, hists, label=label)

f_z = lambda z: 0.3 * np.exp(-0.2 * z ** 2) + 0.7 * np.exp(-0.2 * (z - 10) ** 2)
for scale in [1, 5, 10]:
    t_rvs = lambda x: norm.rvs(loc=x, scale=scale)
    t_pdf = lambda x, loc: norm.pdf(x, loc=loc, scale=scale)
    print('---------- scale: {}-----------'.format(scale))
    test_MH(1200, f_z, t_rvs, t_pdf, 'scale_{}'.format(scale))
plt.legend()
```

*Notes:*
> 1. 对提议转移概率的方差的不同选择会造成不同的结果。
> 2. 如果转移概率形状太狭窄， 则mc链很难遍历各个mode, 如果太宽则可能导致拒绝率过高。
> 3. 如图，scale为1的图形， 由于x0初始化在0附近， 它的方差太小，很难逃出0处的这个mode， 迁移到10处的这个mode， 其它两种选择就避免的这种情况。

## 4. 模拟退火重要性采样(AIS)

```python
# target
f_0 = lambda z: 0.3 * np.exp(-0.2 * z ** 2) + 0.7 * np.exp(-0.2 * (z - 10) ** 2)
f_n = lambda z: np.exp(-(z-5) ** 2 / 4)  # proposal
p_n = norm(loc=5, scale=2)
f_j = lambda z, beta: f_0(z) ** beta * f_n(z) ** (1 - beta)  # intermediate distribution

X = np.arange(-5, 15, 0.1)
for beta in np.arange(0, 1.2, 0.2):
    plt.plot(X, f_j(X, beta), label='f_j_{}'.format(beta))
plt.legend(loc='upper left')
```

```python
def AIS(n_samples, p_n, f_j, betas, it, scale):
    samples = np.zeros(n_samples)
    weights = np.zeros(n_samples)
    for i in range(n_samples):
        w = 1
        x = p_n.rvs()
        for k in range(1, len(betas)):
            for j in range(it):
                x_pre = x + norm.rvs(scale=scale)
                # Acceptance prob
                a = f_j(x_pre, betas[k]) / f_j(x, betas[k])
                if np.random.rand() < a:
                    x = x_pre
            w += np.log(f_j(x, betas[k])) - np.log(f_j(x, betas[k - 1]))

        samples[i] = x
        weights[i] = np.exp(w)
    return samples, weights

n_samples = 500
betas = np.arange(0, 1.1, 0.1)
samples, weights = AIS(n_samples, p_n, f_j, betas, 10, 2)
E_p = 1 / np.sum(weights) * np.sum(weights * samples)
print('期望E_p: {}'.format(E_p))

q_resamples = np.random.choice(samples, len(samples), True, weights/np.sum(weights))
hist2, bins2 = np.histogram(q_resamples, 20)
bins2 = 0.5 * (bins2[1:] + bins2[:-1])
plt.plot(bins2, hist2)
print('期望RE_p: {}'.format(np.mean(q_resamples)))
        

```
